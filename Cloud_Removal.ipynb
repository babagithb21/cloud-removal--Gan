```python
import numpy as np
import random
from os import listdir, mkdir
from os.path import join, exists
import shutil
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageFilter
import warnings
warnings.filterwarnings('ignore')
import os
from perlin_noise import PerlinNoise

# Set the paths for the original (target) and processed (cloudy) images
path_target = '/content/drive/MyDrive/River'  # Original images folder
path_input = '/content/drive/MyDrive/ProcessedImages'  # Processed images (with clouds) folder

# Create the processed images folder if it doesn't exist
if not exists(path_input):
    mkdir(path_input)

# Function to generate Perlin noise for clouds
def generate_perlin_noise(width, height, scale, octaves):
    noise_maker = PerlinNoise(octaves=octaves)
    noise = np.zeros((height, width))
    for i in range(height):
        for j in range(width):
            noise[i][j] = noise_maker([i / scale, j / scale])
    return noise

# Normalize the noise
def normalize_noise(noise):
    min_val = noise.min()
    max_val = noise.max()
    return (noise - min_val) / (max_val - min_val)

# Generate clouds by overlaying multiple layers of Perlin noise
def generate_clouds(width, height, base_scale, octaves, persistence, lacunarity):
    clouds = np.zeros((height, width))
    for octave in range(1, octaves + 1):
        scale = base_scale / octave
        layer = generate_perlin_noise(width, height, scale, octaves)
        clouds += layer * (persistence ** octave)

    clouds = normalize_noise(clouds)
    return clouds

# Overlay clouds on the original image
def overlay_clouds(image, clouds, alpha=0.3, blur_radius=2):
    clouds_rgb = np.stack([clouds] * 3, axis=-1)
    image = image.astype(float) / 255.0
    clouds_rgb = clouds_rgb.astype(float)
    clouds_rgb = Image.fromarray((clouds_rgb * 255).astype(np.uint8))
    clouds_rgb = clouds_rgb.filter(ImageFilter.GaussianBlur(radius=blur_radius))
    clouds_rgb = np.asarray(clouds_rgb) / 255.0

    blended = image * (1 - alpha) + clouds_rgb * alpha
    blended = (blended * 255).astype(np.uint8)
    return blended

# Image size and noise parameters
width, height = 64, 64
octaves = 3
persistence = 0.8
lacunarity = 6

# Get all the .jpg files in the target directory
files_present = [f for f in listdir(path_target) if f.endswith('.jpg')]

# Process each image
for i, filename in enumerate(files_present):
    base_scale = random.uniform(5, 120)
    alpha = random.uniform(0.2, 0.5)
    blur_radius = random.uniform(1, 3)
    clouds = generate_clouds(width, height, base_scale, octaves, persistence, lacunarity)
    img_path = join(path_target, filename)
    img = np.asarray(Image.open(img_path).resize((width, height)))
    image_with_clouds = Image.fromarray(overlay_clouds(img, clouds, alpha, blur_radius))

    # Save the processed image in the 'ProcessedImages' folder
    image_with_clouds.save(join(path_input, filename))

    print(f'Processed {i+1}/{len(files_present)}')

    # Display original and processed images for the first 15 images
    if i < 15:
        fig, ax = plt.subplots(1, 2)
        ax[0].imshow(np.asarray(Image.open(join(path_target, filename))))
        ax[1].imshow(np.asarray(image_with_clouds))
        ax[0].set_title("Original Image")
        ax[1].set_title("Image with Clouds")
        ax[0].axis('off')
        ax[1].axis('off')
        plt.show()
```

```python
import os
from PIL import Image
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
from torchvision import transforms

# 1. Define the Dataset class
class ProcessedToCleanDataset(Dataset):
    def __init__(self, processed_dir, target_dir, transform=None):
        self.processed_dir = processed_dir
        self.target_dir = target_dir
        self.transform = transform

        # Create a dictionary mapping serial numbers to file paths for targets (clean images)
        self.target_images = {self.get_serial_number(f): os.path.join(target_dir, f)
                              for f in os.listdir(target_dir) if f.endswith('.jpg') and self.get_serial_number(f)}

        # Only load processed images that have a corresponding target
        self.processed_images = {self.get_serial_number(f): os.path.join(processed_dir, f)
                                 for f in os.listdir(processed_dir) if f.endswith('.jpg') and self.get_serial_number(f)}

        # Ensure we only load processed images that have matching targets
        self.matched_serials = [serial for serial in self.processed_images if serial in self.target_images]
        print(f"Matched {len(self.matched_serials)} files.")

    def __len__(self):
        return len(self.matched_serials)

    def __getitem__(self, idx):
        serial = self.matched_serials[idx]

        # Paths for the processed (input) and target (clean) images
        processed_path = self.processed_images[serial]
        target_path = self.target_images[serial]

        # Open the images
        processed_image = Image.open(processed_path)
        target_image = Image.open(target_path)

        # Apply transformations if provided (e.g., convert to tensor, normalize, etc.)
        if self.transform:
            processed_image = self.transform(processed_image)
            target_image = self.transform(target_image)

        return processed_image, target_image

    def get_serial_number(self, filename):
        """Extract the serial number from filenames like 'River_<serial no>.jpg'"""
        try:
            return filename.split('_')[1].split('.')[0]
        except IndexError:
            print(f"Error extracting serial number from filename {filename}: Filename does not match expected format")
            return None

# 2. Image Transformations
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Resize to 64x64 (if images aren't already)
    transforms.ToTensor()         # Convert images to tensor format
])

# 3. Directories
processed_dir = '/content/drive/MyDrive/ProcessedImages'  # Directory containing processed images
target_dir = '/content/drive/MyDrive/River'  # Directory containing clean images

# Instantiate the dataset and dataloader
dataset = ProcessedToCleanDataset(processed_dir, target_dir, transform=transform)
train_loader = DataLoader(dataset, batch_size=32, shuffle=False)

# 4. GAN Model Definitions (simplified example)

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        # Define a simple CNN generator architecture
        self.gen = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1),
            nn.Tanh()
        )

    def forward(self, x):
        return self.gen(x)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        # Define a simple CNN discriminator architecture
        self.disc = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Flatten(),
            nn.Linear(128*16*16, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.disc(x)

# 5. Initialize models (use CPU instead of CUDA)
device = torch.device("cpu")

generator = Generator().to(device)
discriminator = Discriminator().to(device)

# Loss and optimizers
bce_loss = nn.BCELoss()
l1_loss = nn.L1Loss()
optimizer_G = torch.optim.Adam(generator.parameters(), lr=0.0002)
optimizer_D = torch.optim.Adam(discriminator.parameters(), lr=0.0002)

# 6. Training loop
num_epochs = 15  # Run for 15 epochs

for epoch in range(num_epochs):
    print(f"Epoch {epoch+1}/{num_epochs}")

    for batch_idx, (inputs, targets) in enumerate(train_loader):
        inputs, targets = inputs.to(device), targets.to(device)
        batch_size = inputs.size(0)

        '''1. Train Discriminator'''
        optimizer_D.zero_grad()

        # Generate fake images
        fake_images = generator(inputs)

        # Real images label = 1, fake images label = 0
        real_labels = torch.ones(batch_size, 1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)

        # Discriminator loss on real images
        outputs_real = discriminator(targets)
        loss_real = bce_loss(outputs_real, real_labels)

        # Discriminator loss on fake images
        outputs_fake = discriminator(fake_images.detach())
        loss_fake = bce_loss(outputs_fake, fake_labels)

        # Combine losses
        loss_D = (loss_real + loss_fake) / 2
        loss_D.backward()
        optimizer_D.step()

        '''2. Train Generator'''
        optimizer_G.zero_grad()

        # Generator loss
        outputs_fake = discriminator(fake_images)
        loss_G_bce = bce_loss(outputs_fake, real_labels)  # Fooling the discriminator
        loss_G_l1 = l1_loss(fake_images, targets) * 100   # L1 loss for pixel-wise accuracy
        loss_G = loss_G_bce + loss_G_l1

        loss_G.backward()
        optimizer_G.step()

        # Print losses
        if batch_idx % 10 == 0:
            print(f"Batch {batch_idx}/{len(train_loader)}: Loss D: {loss_D.item()}, Loss G: {loss_G.item()}")

    ''' Display images at the end of each epoch '''
    with torch.no_grad():
        fig, ax = plt.subplots(1, 3, figsize=(12, 4))

        # Display input image (processed image)
        ax[0].imshow(np.transpose(inputs[0].cpu().numpy(), (1, 2, 0)))
        ax[0].set_title('Processed Image')
        ax[0].axis('off')

        # Display target image (clean image)
        ax[1].imshow(np.transpose(targets[0].cpu().numpy(), (1, 2, 0)))
        ax[1].set_title('Target Image')
        ax[1].axis('off')

        # Display generated image
        ax[2].imshow(np.transpose(fake_images[0].cpu().numpy(), (1, 2, 0)))
        ax[2].set_title('Generated Image')
        ax[2].axis('off')

        plt.tight_layout()
        plt.show()
```

```python
# Import necessary metrics
from skimage.metrics import peak_signal_noise_ratio, structural_similarity
import matplotlib.pyplot as plt
import numpy as np

# Lists to store PSNR and SSIM values across epochs
psnr_values = []
ssim_values = []

# 7. Calculate PSNR and SSIM after each epoch (Using previously generated images)
for epoch in range(num_epochs):
    psnr_epoch = []
    ssim_epoch = []

    for batch_idx, (inputs, targets) in enumerate(train_loader):
        inputs, targets = inputs.to(device), targets.to(device)

        # Generate fake images using the trained generator
        with torch.no_grad():
            fake_images = generator(inputs)

        batch_size = inputs.size(0)

        # Calculate PSNR and SSIM for each image in the batch
        for i in range(batch_size):
            # Convert torch tensors to numpy arrays and transpose them back to HxWxC format
            generated_img = fake_images[i].cpu().detach().numpy().transpose(1, 2, 0)
            target_img = targets[i].cpu().detach().numpy().transpose(1, 2, 0)

            # PSNR and SSIM calculations
            psnr_value = peak_signal_noise_ratio(target_img, generated_img, data_range=1.0)
            ssim_value = structural_similarity(target_img, generated_img, win_size=3,  # Smaller window size for small images
                                               channel_axis=2, data_range=1.0)  # Specify channel axis for RGB

            psnr_epoch.append(psnr_value)
            ssim_epoch.append(ssim_value)

    # Store the average PSNR and SSIM values for the epoch
    avg_psnr = np.mean(psnr_epoch)
    avg_ssim = np.mean(ssim_epoch)
    psnr_values.append(avg_psnr)
    ssim_values.append(avg_ssim)

    print(f"Epoch {epoch+1} - Avg PSNR: {avg_psnr:.4f}, Avg SSIM: {avg_ssim:.4f}")

# 8. Plot PSNR and SSIM Graphs
epochs = list(range(1, num_epochs + 1))

# Plot PSNR graph
plt.figure(figsize=(10, 5))
plt.plot(epochs, psnr_values, label='PSNR', color='b', marker='o')
plt.xlabel('Epochs')
plt.ylabel('PSNR')
plt.title('PSNR over Epochs')
plt.legend()
plt.grid(True)
plt.show()

# Plot SSIM graph
plt.figure(figsize=(10, 5))
plt.plot(epochs, ssim_values, label='SSIM', color='g', marker='o')
plt.xlabel('Epochs')
plt.ylabel('SSIM')
plt.title('SSIM over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
